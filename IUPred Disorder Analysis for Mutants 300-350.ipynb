```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sequence = pd.read_csv("alanine_scanning_IUPred.csv", sep = '	',names = ['#POS','AMINO ACID','IUPRED SCORE','ANCHOR SCORE'])
nonmutant_sequence = sequence.iloc[6089:6590]
nonmutant_sequence.head()
```

```python
nonmutant_sequence['#POS'] = nonmutant_sequence['#POS'].astype(float)
nonmutant_sequence['IUPRED SCORE'] = nonmutant_sequence['IUPRED SCORE'].astype(float)
```

```python
mutant317_sequence = sequence.iloc[5:506]
mutant317_sequence.head()
```

```python
mutant317_sequence['#POS'] = mutant317_sequence['#POS'].astype(float)
mutant317_sequence['IUPRED SCORE'] = mutant317_sequence['IUPRED SCORE'].astype(float)
```

```python
nonmutant_plot = nonmutant_sequence.plot(x = '#POS', y = 'IUPRED SCORE',label='No mutation')
mutant317_plot = mutant317_sequence.plot(ax=nonmutant_plot,x = '#POS', y = 'IUPRED SCORE',title = 'Mutation at residue 317',label='Mutation at residue 317')
mutant317_plot.set_xlabel('Residue #')
mutant317_plot.set_ylabel('Disorder Score')
```
